```python
#implimenting multiple linear regression
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("company_.csv")
df
```

```python
df.head(5)
```

```python
df.info()
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
print(x.shape)
print(y.shape)
```

Scatter plots are drawn to display the relationship between individual features and the target variable.

```python
x1 = df.iloc[:,0]
y = df.iloc[:,-1]
plt.scatter(x1,y)
plt.show()
```

```python
x2 = df.iloc[:,1]
y = df.iloc[:,-1]
plt.scatter(x2,y)
plt.show()
```

```python
x3 = df.iloc[:,2]
y = df.iloc[:,-1]
plt.scatter(x3,y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.25,random_state=1)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

```python
regressor.fit(xtrain,ytrain)
```

```python
predictions= regressor.predict(xtest)
results = pd.DataFrame({"Actual":ytest,"Predictions":predictions})
results
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_absolute_error(ytest,predictions)
```

```python
mean_squared_error(ytest,predictions)
```

```python
r2_score(ytest,predictions)
```
